```python
import json
import matplotlib.pyplot as plt
```

```python
filename = 'MICROarray+.json'

with open(filename) as f:
    data = json.load(f)['results']
```

```python
def plotLine(onx, ony, pa, page):
    filtered = [r for r in data if r['pa']==pa and r['isPage']==page]
    x = [float(r[onx]) for r in filtered]
    y = [float(r[ony]) for r in filtered]
    
#     x = x[1:]
#     y = y[1:]

    plt.plot(x, y, '-ro')
    algo = 'Const Stride' if pa == 0 else 'LinkedList'
    pageblock = 'Block' if page == 0 else 'Page'
    plt.xlabel(onx)
    plt.ylabel(ony)
    plt.title('{0} vs {1} - {2} - {3}'.format(onx, ony, algo, pageblock))
    plt.xticks(x)
    plt.ylim(0-0.05, 1+0.05)
    plt.xlim(min(x), max(x)+5)
    plt.grid(color='grey', linestyle='-', linewidth=0.1)
    plt.show()

# Array
plotLine('pd', 'coverage', 0,0)
plotLine('pd', 'misfetchrate', 0,0)
plotLine('pd', 'coverage', 0,1)
plotLine('pd', 'misfetchrate', 0,1)

# Linkedlist
plotLine('pd', 'coverage', 1,0)
plotLine('pd', 'misfetchrate', 1,0)
plotLine('pd', 'coverage', 1,1)
plotLine('pd', 'misfetchrate', 1,1)
```
